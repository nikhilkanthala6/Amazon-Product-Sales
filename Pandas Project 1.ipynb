```python
num_df = df.select_dtypes(include = 'number')
cat_df = df.select_dtypes(exclude = 'number')
```

```python
for column in num_df.columns:
    num_df[column] = df[column].fillna(df[column].median())
```

```python
num_df.isnull().sum()
```

```python
for column in cat_df.columns:
    cat_df[column] = cat_df[column].fillna(cat_df[column].mode()[0])
```

```python
cat_df.isnull().sum()
```

```python
df = pd.concat([num_df, cat_df], axis =1)
```

```python
df.head()
```

```python
df['total_reviews'] = df['total_reviews'].astype(int)
df['purchased_last_month'] = df['purchased_last_month'] .astype(int)
```

```python
df.info()
```

```python
df['delivery_date'] = pd.to_datetime(df['delivery_date'], errors = 'coerce')
df['data_collected_at'] = pd.to_datetime(df['data_collected_at'], errors = 'coerce')
```

```python
df.info()
```

```python
df.drop(['sustainability_tags'], axis = 1, inplace = True)
```

```python
df.head(5)
```

## Feature Engineering

```python
df ['Discounted_Amount'] = df ['original_price'] - df ['discounted_price']
```

```python
df.head(5)
```

```python
df['Total_revenue'] = df ['discounted_price'] * df ['purchased_last_month']
```

```python
df ['percentage_band'] = pd.cut (x = df['discount_percentage'],
                                bins = [0,15,45,65,99],
                                labels = ['Less discount', 'Medium', 'High', 'Very high'])
```
